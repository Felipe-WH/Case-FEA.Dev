```python
map_dictionary = {"A" : 0, "B" :1 , "C": 2 , "D":3, "E":4, "F":5, "G":6}                         # Crio um dicionário para fazer essa representação numérica das letras, sendo A, menor número, e G, maior número
filtro = df.drop(["person_home_ownership","loan_intent", "cb_person_default_on_file"],axis=1)    # Crio um filtro do dataframe com algumas colunas de interesse
filtro["loan_grade"] = filtro["loan_grade"].map(map_dictionary)                                  # Utilizo a função map() para trocar os valores da coluna "loan_grade"

colunas = df.drop(["person_home_ownership","loan_intent", "cb_person_default_on_file", "loan_grade"],axis=1).columns   # Variável "colunas" com o valor das colunas de interesse
correlacao = []            # Variável "correlacao" como sendo uma lista vazia
for i in colunas:          # Loop para rodar a lista colunas
    correlacao.append(round(filtro[[i,"loan_grade"]].corr().iloc[0]["loan_grade"],4))    # Adiciona na lista o valor da correlação da coluna do loop pela coluna mapeada de "loan_grade"
    
df1= pd.DataFrame({"Correlacao x Score": correlacao})             # Crio um dataframe para melhor vizualização dos dados adiquiridos
df1.index = colunas        # Define o index do dataframe criado como sendo a variável "colunas"
df1 = df1.sort_values(by="Correlacao x Score",ascending=False)    # ORdena os valores do dataframe por ordem decrescente
  
plt.figure(figsize=(20,9))                                        # Define tamanho do plot do gráfico
pl = plt.bar(df1.index,df1["Correlacao x Score"], width=0.8)      # Cria gráfico de barra com os valores do dataframe criado
plt.rc('xtick', labelsize=15)                                     # Define o tamanho das letras do label x
plt.rc('ytick', labelsize=20)                                     # Define o tamanho das letras do label y
plt.title("Correlação entre as variáveis e o Score", fontsize=20) # Define o nome do título do gráfico


for bar in pl:     # Esse for serve apenas para exibir em cima dos gráficos o valor correspondente da barra
        plt.annotate(f"{round(bar.get_height(),4)}", xy=(bar.get_x()+0.05, bar.get_height()+0.04), fontsize=20)
        
display(df1)       # Exbibe o dataframe criado
```

## Exercício 6: Considerando as subdivisões apresentadas anteriormente, responda:

Analise a quantidade total de créditos com default e os bons pagadores, mostre a porcentagem que cada grupo representa no total

Qual o tipo de pessoa que tem maior probabilidade de dar calote?

Dicas:
  - Procure pelo método groupby
  - Procure pelo método pivot table
  - Não é necessário utilizá-los porém é um caminho mais fácil


<font color=pink|black|black>*###SUA RESPOSTA COMECA AQUI###*</font>

```python
#groupby loan_grade soma loan status
resumo_grade = df[["loan_grade", "loan_status"]].groupby("loan_grade").sum()                           # Atribui à variável, um groupby(), ou seja, um agrupamento da coluna "loan_grade" e a soma dos valores respectivos da coluna "loan_status"
resumo_intent = df[["loan_intent", "loan_status"]].groupby("loan_intent").sum()                        # Atribui à variável, um groupby(), ou seja, um agrupamento da coluna "loan_intent" e a soma dos valores respectivos da coluna "loan_status"
resumo_home = df[["person_home_ownership", "loan_status"]].groupby("person_home_ownership").sum()      # Atribui à variável, um groupby(), ou seja, um agrupamento da coluna "loan_person_home_ownership" e a soma dos valores respectivos da coluna "loan_status"
resumo_person_length = df[["person_emp_length", "loan_status"]].groupby("person_emp_length").sum()     # Atribui à variável, um groupby(), ou seja, um agrupamento da coluna "person_emp_length" e a soma dos valores respectivos da coluna "loan_status"


for i in [resumo_grade, resumo_intent, resumo_home, resumo_person_length]:    # For dentro das variáveis de groupby 
    i["%"] = round(i["loan_status"]/i["loan_status"].sum()*100,2)             # Adiciona uma coluna "%" para representar a proporção de cada valor da coluna "loan_status"

resumo_grade = resumo_grade.sort_values(by ="loan_status", ascending = False)                   # Ordena as variáveis do maior "%" para o menor
resumo_intent = resumo_intent.sort_values(by ="loan_status", ascending = False)                 # Ordena as variáveis do maior "%" para o menor
resumo_home = resumo_home.sort_values(by ="loan_status", ascending = False)                     # Ordena as variáveis do maior "%" para o menor
resumo_person_length = resumo_person_length.sort_values(by ="loan_status", ascending = False)   # Ordena as variáveis do maior "%" para o menor


display(resumo_grade,resumo_intent,resumo_home,resumo_person_length)                                   # Exibe todos os groupby calculados
variaveis = [resumo_grade,resumo_intent,resumo_home,resumo_person_length]                              # Atribui à variável "variaveis", uma lista com todas as variaveis de groupby

colunas = ["loan_grade","loan_intent","person_home_ownershop","person_emp_length"]     # Atribui à variável uma lista com as colunas de interesse
vet= []          # Atribui à variável "vet" uma lista vazia

for i in variaveis:     # Loop para rodar dentro da lista "variaveis"
    vet.append(round(i["loan_status"].max()/i["loan_status"].sum()*100,2))    # Adiciona na lista criada o percentual de calotes (%) de cada valor total para cada um dos groupby
    
    
calote_df=[]     # Atribui à variável "calote_df" uma lista vazia
for i in variaveis:      # Loop para rodar dentro da lista "variaveis"
    i["coluna"]= i.index                                                       # Para cada groupby, é criado uma nova coluna auxiliar
    x = i["coluna"][i["loan_status"]==i["loan_status"].max()].max()            # Atribui à variável "x", um filtro apenas do primeiro valor de maior participação (%) de cada groupby
    calote_df.append(x)                                                       # Adiciona esses valores na lista "calote_df"

    
Calotes = pd.DataFrame({"Porcentagem de calotes (%)":vet,"Principais grupos de calote":calote_df})    # Cria um novo DataFrame para melhor exibição dos dados              
Calotes.index = colunas    # Define o index do DF como sendo a lista "colunas"
display(Calotes)           # Exibe DF
                         
```
